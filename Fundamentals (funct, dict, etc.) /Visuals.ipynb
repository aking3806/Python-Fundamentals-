##### Change Ticks and Lables

```python
##Manipulate xlabels and ticks formatting 
fig = plt.figure(figsize=(5,3))
ax = plt.axes((0.1,0.1,0.5,0.8))

#change ticks on figure
ax.tick_params(axis='x',
               which='major',
               direction='out',
               length=10, #length of ticks
               width=3, #width of ticks
               color='black', #color of ticks
               pad=10, #space between ticks
               labelsize=10, #label size
               labelcolor='black', #color of x label
               labelrotation=10) #rotation of x label
plt.show()
```

```python
### Manipulate number of x ticks and labels 
fig = plt.figure(figsize=(5,3))
ax = plt.axes((0.1,0.1,0.5,0.8))

#set to only two x ticks
ax.set_xticks([0.2,1])

#label x ticks
ax.set_xticklabels(['pos1','pos2'])
plt.show()
```

```python
#Change Display of tick Lables 
fig = plt.figure(figsize=(5,3))
ax = plt.axes((0.1,0.1,0.5,0.8))

#adjust tick labels 
ax.yaxis.set_major_locator(MultipleLocator(0.5)) ##set y axis major to show at .5 
ax.yaxis.set_minor_locator(MultipleLocator(0.1)) ##set y axis major for every .1 

#change gridlines 
ax.grid(True,which='major',axis='y',alpha=1)

plt.show()
```

#### Legend 

shadow: [None or bool] Whether to draw a shadow behind the legend.Its Default value is None.
markerscale: [None or int or float] The relative size of legend markers compared with the originally drawn ones.The Default is None.
numpoints: [None or int] The number of marker points in the legend when creating a legend entry for a Line2D (line).The Default is None.
fontsize: The font size of the legend.If the value is numeric the size will be the absolute font size in points.
facecolor: [None or inherit or color] The legends background color.
edgecolor: [None or inherit or color] The legends background patch edge color.

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.legend(['line'], bbox_to_anchor =(1, 1.15), ncol = 2)
```
