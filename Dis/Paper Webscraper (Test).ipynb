```python
import requests
from bs4 import BeautifulSoup
import csv
import datetime
```

```python
url = 'https://arxiv.org/list/cs.HC/pastweek?skip=0&show=25'
```

Step 3: Define the function to extract the research paper information

```python
def scrape_arxiv(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    paper_list = soup.find_all('div', class_='arxiv-result')
    
    papers = []
    
    for paper in paper_list:
        title = paper.find('p', class_='title is-5 mathjax').text.strip()
        authors = paper.find('p', class_='authors').text.strip()[9:] # Remove "Authors: " from the beginning
        abstract = paper.find('span', class_='abstract-full has-text-grey-dark mathjax').text.strip()
        link = "https://arxiv.org" + paper.find('a', title='Abstract')['href']
        
        papers.append({"Paper Title": title, "Authors": authors, "Abstract": abstract, "Link to the Paper": link})
        
    return papers

print(scrape_arxiv(url))
```
