**Using pandas query function**
- Can be used with and, or, 
   or: df.query("A > 2 or F > 4")
   and: df.query("A > 2 and E == 5")
- can include 'not' for opposite: 
   df.query("not (A > 2 or F > 4)")
- can also use math operators: 
   df.query("A*2 < 5")
- use for dates as well: 
  df.query("Date.dt.month > 6") -- where month is > 6


```python
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.randint(10, size=(5,6)), columns=list("ABCDEF"))
df
```

```python
#only rows where col A > 4 
df.query("A > 4")
```

```python
#only rows where col A > 4  & E = 5
df.query("A > 4 and E == 5")
```

```python
#only rows where col A > 2  or F > 4
df.query("A > 2 or F > 4")
```
